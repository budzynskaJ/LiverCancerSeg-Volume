```python
import pandas as pd
import matplotlib.pyplot as plt

column_names = ["Liczba iteracji", "Wątroba (Dokładność)", "Guz (Dokładność)", "Wątroba (IoU)", "Guz (IoU)"]

resnest = pd.read_csv('/home/budzynskaj/project/LiverCancerProject/LiverCancerSeg-Volume/resnest-val2.txt', sep=',', header=None, names=column_names)

resnest = resnest.apply(pd.to_numeric)


print(resnest)
```

```python
import matplotlib.pyplot as plt

# Create line charts for accuracy and IoU of each organ
plt.figure(figsize=(10, 6))

# Wątroba
plt.plot(resnest["Liczba iteracji"], resnest["Wątroba (Dokładność)"], label="Wątroba (Dokładność)", color='dodgerblue')
plt.plot(resnest["Liczba iteracji"], resnest["Wątroba (IoU)"], label="Wątroba (IoU)", color = 'dodgerblue', linestyle='--')

# Guz
plt.plot(resnest["Liczba iteracji"], resnest["Guz (Dokładność)"], label="Guz (Dokładność)", color='orangered')
plt.plot(resnest["Liczba iteracji"], resnest["Guz (IoU)"], label="Guz (IoU)", color='orangered', linestyle='--')

# Add labels and title
plt.xlabel("Liczba iteracji")
plt.ylabel("Wartość w %")
plt.title("Dokładność i IoU w zależności od liczby iteracji dla architektury ResNeSt")

# Add legend
plt.legend()

# Show the plot
plt.show()
```
