```python
import mmcv
import matplotlib.pyplot as plt
import numpy as np

path = "/home/budzynskaj/project/LiverCancerProject/data/d/dane/volumes/volume-28_73.png"
equalized_path = "/home/budzynskaj/project/LiverCancerProject/data/d/dane/volumes_norm_final/volume-28_73.png"
maska_path = "/home/budzynskaj/project/LiverCancerProject/data/d/dane/volume_labels/segmentation-28_73.png" 
image = mmcv.imread(path, flag='grayscale')
equalized_image = mmcv.imread(equalized_path, flag='grayscale')
mask = mmcv.imread(maska_path)

# Convert grayscale images to RGB for consistent plotting
image_rgb = mmcv.bgr2rgb(mmcv.gray2rgb(image))
equalized_image_rgb = mmcv.bgr2rgb(mmcv.gray2rgb(equalized_image))

# Alpha blending
alpha = 0.3  
overlay_image = image_rgb.copy()


mask = mmcv.bgr2rgb(mask)

overlay_image = (1 - alpha) * overlay_image + alpha * mask

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(image_rgb, cmap='gray')
plt.title('Obraz oryginalny')

plt.subplot(1, 3, 2)
plt.imshow(equalized_image_rgb, cmap='gray')
plt.title('Obraz po wyrównaniu histogramu')

plt.subplot(1, 3, 3)
plt.imshow(overlay_image.astype(np.uint8))
plt.title('Obraz z nałożoną maską')

plt.show()
```
