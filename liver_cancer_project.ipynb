```python
import os

def przenies_zdjecia(folder_wejsciowy, folder_wyjsciowy):

    if not os.path.exists(folder_wejsciowy) or not os.path.exists(folder_wyjsciowy):
        
        return
    
    pliki = os.listdir(folder_wejsciowy)

    for plik in pliki:
        sciezka_pliku = os.path.join(folder_wejsciowy, plik)

        if os.path.isfile(sciezka_pliku) and plik.startswith("segmentation-") and "livermask" in plik:
            
            nowa_sciezka = os.path.join(folder_wyjsciowy, plik)

            os.rename(sciezka_pliku, nowa_sciezka)
        
# TODO poprawi cieki na swoje
folder_wejsciowy = "C:/Users/Uytkownik/Desktop/studia/MAGISTERKA/II sem/Projekt badawczy/projekt/data/dataset_6"
folder_wyjsciowy = "C:/Users/Uytkownik/Desktop/studia/MAGISTERKA/II sem/Projekt badawczy/projekt/data/labels"

przenies_zdjecia(folder_wejsciowy, folder_wyjsciowy)
```
